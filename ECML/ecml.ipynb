```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
from deepforest import CascadeForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import pickle
```

```python
path = 'training_set/params_train/0205_01_05.txt'
def read_one_params_data(path):
    with open(path) as f:
        data = np.loadtxt(path, delimiter='\t', dtype=np.double)
    name = path.split('/')[-1].split(".")[0]
    return name, data
read_one_params_data(path)
```

```python
def walkFile(file, mode): # 若内存不足，需修改成读部分数据
    for root, dirs, files in os.walk(file):
        # root 表示当前正在访问的文件夹路径
        # dirs 表示该文件夹下的子目录名list
        # files 表示该文件夹下的文件list
        name_list, data_list = [], []
        # 遍历文件
        for f in tqdm(files):
            path = os.path.join(root, f)
            if mode == "noisy":
                name, data = read_one_nosiy_data(path)
            elif mode == "params":
                name, data = read_one_params_data(path)
            name_list.append(name)
            data_list.append(data)
    return name_list, np.array(data_list)
name_list, y_data_list = walkFile("./home/ucapats/Scratch/ml_data_challenge/training_set/params_train","params")
df_y = pd.DataFrame(y_data_list)
df_y["name"] = name_list
df_y
```

```python
path = './home/ucapats/Scratch/ml_data_challenge/training_set/noisy_train/0455_08_04.txt'
def read_one_nosiy_data(path):
    with open(path) as f:
        content = f.read()
        content = content.replace("# star_temp: ","").replace("# star_logg: ","").replace("# star_rad: ","").replace("# star_mass: ","").replace("# star_k_mag: ","").replace("# period: ","")
        content = content.replace("\n", "\t")
        data = np.fromstring(content, sep='\t', dtype=np.double)
    name = path.split('/')[-1].split(".")[0]
    return name, data
name, data = read_one_nosiy_data(path)
data.shape
```

```python
df_X_file = "df_X.pk"
if os.path.exists(df_X_file):
    with open(df_X_file, "rb") as f:
        df_X = pickle.load(f)
else:
    name_list, data_list = walkFile("./home/ucapats/Scratch/ml_data_challenge/training_set/noisy_train", "noisy")
    df_X = pd.DataFrame(data_list)
    df_X["name"] = name_list
    with open(df_X_file, "wb") as f:
        pickle.dump(df_X, f, protocol = 4)
    df_X
```

```python
df_test_file = "df_test.pk"
if os.path.exists(df_test_file):
    with open(df_test_file, "rb") as f:
        df_test = pickle.load(f)
else:
    name_list, data_list = walkFile("./home/ucapats/Scratch/ml_data_challenge/test_set/noisy_test", "noisy")
    df_test = pd.DataFrame(data_list)
    df_test["name"] = name_list
    with open(df_test_file, "wb") as f:
        pickle.dump(df_test, f, protocol = 4)
    df_test
```

```python
final_mode = False
if final_mode:
    x_train, y_train, x_test = df_X, df_y, df_test
else:
    x_train, x_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.99) # 只用1%数据训练
x_train.shape
```

```python
DF21 = CascadeForestRegressor(n_jobs=20)
DF21.fit(x_train.drop(columns=['name']).values, y_train.drop(columns=['name']).values)
```

```python
y_pred = DF21.predict(x_test.drop(columns=['name']).values)
if final_mode:
    df_output = pd.DataFrame(y_pred)
    df_output["name"] = x_test["name"]
    df_output = df_output.sort_values("name")
    df_output.drop(columns=['name']).to_csv("output.txt", sep='\t',float_format='%.12f', header=False, index=False)
else:
    mse = mean_squared_error(y_test.drop(columns=['name']).values, y_pred)
    print("\nTesting MSE: {:.9f}".format(mse))
```
